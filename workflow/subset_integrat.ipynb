```python
from matplotlib import pyplot as plt
import networkx as nx

G = nx.DiGraph()

G.add_node("Read h5ad Files Concurrently")
G.add_node("Quality Control")
G.add_node("Data Normalization")
G.add_node("Hierarchical Subsetting")
G.add_node("Harmony Integration")
G.add_node("Downstream Analysis")

G.add_edges_from([
    ("Read h5ad Files Concurrently", "Quality Control"),
    ("Quality Control", "Data Normalization"),
    ("Data Normalization", "Hierarchical Subsetting"),
    ("Hierarchical Subsetting", "Harmony Integration"),
    ("Harmony Integration", "Downstream Analysis"),
])

pos = nx.spring_layout(G)
plt.figure(figsize=(12, 7))
nx.draw(G, pos, with_labels=True, node_size=2500, node_color="lightblue", font_size=10, font_weight="bold")
plt.title("Workflow for Handling h5ad Files with Hierarchical Subsetting and Harmony Integration")
plt.show()
```
