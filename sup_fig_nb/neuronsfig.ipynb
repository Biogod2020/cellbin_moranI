```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import wilcoxon

# Load the data
df = pd.read_csv('./inn_moranI.csv')

# Define the significance threshold for the Wilcoxon test
significance_level = 0.05

# Get unique regions
regions = df['region_abb'].unique()

# Set up the figure with subplots (1 row, number of unique regions)
fig, axes = plt.subplots(1, len(regions), figsize=(20, 6), sharey=True)

# Loop through each region and plot
for i, region in enumerate(regions):
    # Filter data for each region
    data = df[df['region_abb'] == region]
    
    # Filter data to include only the 'fine' values in real_celldict for the current region
    data = data[data['fine'].isin(real_celldict[region])]
    
    # Set x-axis limits
    x_min, x_max = -0.005, 0.025
    
    # Filter data to include only points within the x-axis limits
    data_filtered = data[(data["Moran's I"] >= x_min) & (data["Moran's I"] <= x_max)]
    
    # Create a boxplot for each subplot with alpha set to 0.7
    sns.boxplot(x="Moran's I", y='fine', data=data, ax=axes[i], palette=cell_type_colors, boxprops=dict(alpha=0.7), fliersize=0,
                linewidth=0.5)
    
    # Add a strip plot for each box plot, using the filtered data
    sns.stripplot(x="Moran's I", y='fine', data=data_filtered, ax=axes[i], palette=cell_type_colors, alpha=0.5, jitter=True, 
                  edgecolor='grey', linewidth=0.2)
    
    # Add a red vertical dashed line at x=0
    axes[i].axvline(x=0, color='red', linestyle='--')
    
    # Set the title for each subplot
    axes[i].set_title(region)
    
    # Set x-axis limits
    axes[i].set_xlim(x_min, x_max)
    
    # Loop through each category ('fine') in the region
    for j, fine_category in enumerate(data['fine'].unique()):
        # Extract Moran's I values for this fine category
        morans_i_values = data[data['fine'] == fine_category]["Moran's I"]
        
        # Check if there are enough values to perform the test (Wilcoxon requires at least 2)
        if len(morans_i_values) > 1:
            # Perform a one-sample Wilcoxon signed-rank test against 0
            stat, p_value = wilcoxon(morans_i_values - 0, alternative='greater')
            
            # Check if the p-value is less than the significance level
            if p_value < significance_level:
                # Add an asterisk to the right tick of the box corresponding to this fine category
                axes[i].text(0.02, j, '*', color='red', fontsize=20, va='center')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.savefig('/Users/jay/LocalProjects/cellbin_moranI/sup_fig_nb/boxplot_moranI_inn.pdf', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
from scipy import stats
from scipy.stats import wilcoxon

exn_df = pd.read_csv('ExN_dist_moranI.csv')

# Define the significance threshold for the Wilcoxon test
significance_level = 0.05

# Get unique regions
regions = exn_df['region_abb'].unique()

# Set up the figure with subplots (1 row, number of unique regions), sharey=True to share y-axis
fig, axes = plt.subplots(1, len(regions), figsize=(20, 6), sharey=True)

# Define the fixed y-limits and x-limits ratio
fixed_y_lim = (14.5, -0.5)
fixed_y_range = fixed_y_lim[0] - fixed_y_lim[1]  # y-range = 15.0
fixed_x_lim = (-0.005, 0.02)
fixed_x_range = fixed_x_lim[1] - fixed_x_lim[0]  # x-range = 0.025

# Loop through each region and plot
for i, region in enumerate(regions):
    # Filter data for each region
    data = exn_df[exn_df['region_abb'] == region]
    
    # Filter data to include only the 'fine' values in real_celldict for the current region
    data = data[data['fine'].isin(real_celldict[region])]
    
    # Create a boxplot for each subplot with alpha set to 0.7
    sns.boxplot(x="Moran's I", y='fine', data=data, ax=axes[i], palette=cell_type_colors, boxprops=dict(alpha=0.7), fliersize=0)
    
    # Add a strip plot for each box plot
    sns.stripplot(x="Moran's I", y='fine', data=data, ax=axes[i], palette=cell_type_colors, alpha=0.5, jitter=True, 
                  edgecolor='grey', linewidth=1)
    
    # Set the title for each subplot
    axes[i].set_title(region)
    
    # Calculate the dynamic x-limits based on the data
    current_x_lim = axes[i].get_xlim()
    current_x_range = current_x_lim[1] - current_x_lim[0]  # Get the current x-range
    
    # Calculate scaling factor based on the fixed y:x ratio
    scale_factor = current_x_range / fixed_x_range
    
    # Adjust the x-limits to maintain the same y:x ratio
    new_x_lim = (current_x_lim[0], current_x_lim[0] + scale_factor * fixed_x_range)
    axes[i].set_xlim(new_x_lim)

    # Set the fixed y-limits for all subplots (because sharey=True, this is applied to all)
    axes[i].set_ylim(fixed_y_lim)

    # Add a red vertical dashed line at x=0
    axes[i].axvline(x=0, color='red', linestyle='--')
    
    # Loop through each category ('fine') in the region
    for j, fine_category in enumerate(data['fine'].unique()):
        # Extract Moran's I values for this fine category
        morans_i_values = data[data['fine'] == fine_category]["Moran's I"]
        
        # Check if there are enough values to perform the test (Wilcoxon requires at least 2)
        if len(morans_i_values) > 1:
            # Perform a one-sample Wilcoxon signed-rank test against 0
            stat, p_value = wilcoxon(morans_i_values - 0, alternative='greater')
            
            # Check if the p-value is less than the significance level
            if p_value < significance_level:
                # Add an asterisk to the right tick of the box corresponding to this fine category
                axes[i].text(0.017, j, '*', color='red', fontsize=20, va='center')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.savefig('/Users/jay/LocalProjects/cellbin_moranI/sup_fig_nb/exn_boxplot_moranI_freex.pdf', dpi=300)
plt.show()
```
