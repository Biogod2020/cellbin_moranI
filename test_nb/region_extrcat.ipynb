```python
import pandas as pd




# Define groups and their related structures
groups = {
    "PFC": ["Secondary motor area"],
    "STR": ["Striatum"],
    "ECT": ["Ectorhinal area"],
    "HPF": ["Hippocampal formation"],
    "TH": ["Thalamus"],
    "RS": ["Retrosplenial area"],
    "BF": ["Substantia innominata", "Magnocellular nucleus", "Diagonal band nucleus", "Hypothalamus"]
}

```

```python
df = pd.read_csv("/Users/jay/OneDrive/桌面/嗑盐/Yu's Lab/allen_ontology.csv")
```

```python
df.to_csv("allen_ontology.csv", index=False)
```

```python
# Create a dictionary to hold the mappings
safe_name_to_group = {}

# Iterate through the DataFrame and match structure_id_path to groups
for idx, row in df.iterrows():
    safe_name = row['safe_name']
    structure_id_path = row['structure_id_path']
    
    # Check each group
    for group, structures in groups.items():
        for structure in structures:
            # Check if the structure_id_path contains the structure's safe_name
            if any(structure_id_path.startswith(str(df[df['safe_name'] == s]['structure_id_path'].values[0])) for s in structures if not df[df['safe_name'] == s].empty):
                safe_name_to_group[safe_name] = group

# Print the resulting dictionary
safe_name_to_group
```
